## 희소 벡터 : 원-핫 인코딩

```python
import torch

dog = torch.FloatTensor([1, 0, 0, 0, 0])
cat = torch.FloatTensor([0, 1, 0, 0, 0])
cow = torch.FloatTensor([0, 0, 1, 0, 0])
pig = torch.FloatTensor([0, 0, 0, 1, 0])
duck = torch.FloatTensor([0, 0, 0, 0, 1])
```

```python
print(torch.cosine_similarity(dog, cat, dim=0))
print(torch.cosine_similarity(dog, cow, dim=0))
print(torch.cosine_similarity(dog, pig, dim=0))
print(torch.cosine_similarity(dog, duck, dim=0))
```
