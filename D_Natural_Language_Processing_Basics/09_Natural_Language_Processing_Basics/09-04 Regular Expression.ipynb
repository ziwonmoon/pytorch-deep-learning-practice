# 정규 표현식 모듈 함수 예제

## re.match() vs re.search()

```python
r = re.compile("ab.")
r.match("kkkabc")
```

```python
r.search("kkkabc")
```

```python
r.match("abckkk")
```

## re.split()

```python
text = "사과 딸기 수박 메론 바나나"
re.split(" ", text)
```

```python
text = """사과
딸기
수박
메론
바나나"""
re.split("\n", text)
```

```python
text = "사과+딸기+수박+메론+바나나"
re.split("\+", text)
```

```python
text = """이름 : 김철수
전화번호 : 010-1234-5678
주소 : 서울시 강남구 역삼동
나이 : 30
성별 : 남성"""

re.findall("\d+", text)
```

```python
re.findall("\d+", "뭉자열이다")
```

```python
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub("[^a-zA-Z]", " ", text)
print(preprocessed_text)
```

# 정규 표현식을 이용한 토큰화

```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the name, Mr. John Doe isn't an actual Mr., but a placeholder name in legal actions, case studies, or when the true identity of a person is unknown or must be withheld for legal reasons."

tokenizer1 = RegexpTokenizer("[\w]+")
tokenizer2 = RegexpTokenizer("\s+", gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))    
```
