# Bag of Words

## '정부가 발표하는 물가상승률과 소비자가 느끼는 물가상승률은 다르다'라는 문장을 BoW로 만들기

```python
from konlpy.tag import Okt

okt = Okt()

def build_bag_of_words(document):
    documnet = document.replace('.', '')
    tokenized_document = okt.morphs(document)

    word_to_index = {}
    bow = []

    for word in tokenized_document:
        if word not in word_to_index.keys():
            word_to_index[word] = len(word_to_index) # 단어가 처음 나왔을 때, 단어의 인덱스는 현재까지 등장한 단어의 수
            bow.insert(len(word_to_index) - 1, 1) # 단어가 처음 나왔을 때, 단어의 인덱스에 해당하는 위치에 1을 삽입
        else:
            index = word_to_index[word] # 단어가 이미 나왔을 때, 단어의 인덱스
            bow[index] += 1 # 단어가 이미 나왔을 때, 단어의 인덱스에 해당하는 위치에 1을 더함

    return word_to_index, bow
```

```python
doc1 = "정부가 발표하는 물가상승률과 소비자가 느끼는 물가상승률은 다르다."
vocab, bow = build_bag_of_words(doc1)
print('vocab : ', vocab)
print('bow : ', bow)
```

'소비자는 주로 소비하는 상품을 기준으로 물가상승률을 느낀다'

```python
doc2 = '소비자는 주로 소비하는 상품을 기준으로 물가상승률을 느낀다.'

vocab, bow = build_bag_of_words(doc2)
print('vocab : ', vocab)
print('bow : ', bow)
```

'정부가 발표하는 물가상승률과 소비자가 느끼는 물가상승률은 다르다. 소비자는 주로 소비하는 상품을 기준으로 물가상승률을 느낀다.'

```python
doc3 = doc1 + ' '  + doc2
vocab, bow = build_bag_of_words(doc3)
print('vocab : ', vocab)
print('bow : ', bow)
```

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

courpus = ['you know I want your love. because I love you.']
vector = CountVectorizer()

print('bow : ', vector.fit_transform(courpus).toarray())
print('vocabulary : ', vector.vocabulary_)
```

1. CountVectorzier는 띄어쓰기만을 기준으로 토큰화를 한다. 한국어에는 적합하지 않다.
2. 길이가 2이상인 문자에 대해서만 토큰으로 인식한다

## 불용어를 제거한 BoW만들기

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
text = ["Family is not an important thing. It's everything."]
vect = CountVectorizer(stop_words=["the", "a", "an", "is", "not"])
print('bow : ', vect.fit_transform(text).toarray())
print('vocabulary : ', vect.vocabulary_)
```

```python
text = ["Family is not an important thing. It's everything."]
vect = CountVectorizer(stop_words=stopwords.words('english')) # NLTK에서 제공하는 영어 불용어 리스트를 사용하여 불용어 제거
print('bow : ', vect.fit_transform(text).toarray())
print('vocabulary : ', vect.vocabulary_)
```

```python
text = ["Family is not an important thing. It's everything."]
vect = CountVectorizer(stop_words="english") # CountVectorizer에서 제공하는 영어 불용어 리스트를 사용하여 불용어 제거
print('bow : ', vect.fit_transform(text).toarray())
print('vocabulary : ', vect.vocabulary_)
```
